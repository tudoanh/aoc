```python
import logging
import sys

logging.basicConfig(stream=sys.stdout, level=logging.INFO)

# Test
logger = logging.getLogger('AOC')
```

```python
universe_size = (len(inp) + 2, len(inp[0]) + 2)
seed_position = (1, 1)
universe = np.full(universe_size, np.NaN)
x_start, y_start = seed_position[0], seed_position[1]

new_inp = []
for l in inp:
    temp = []
    for i in l:
        if i == 'L':
            temp.append(0)
        if i == '.':
            temp.append(np.NaN)
        if i == '#':
            temp.append(1)
    new_inp.append(temp)
seed_array = np.array(new_inp)
x_end, y_end = x_start + seed_array.shape[0], y_start + seed_array.shape[1]
universe[x_start:x_end, y_start:y_end] = seed_array
```

```python
def survival(x, y, universe):
    cell = universe[x, y]
    if cell == np.NaN:
        return cell
    neighbours = universe[x - 1 : x + 2, y - 1 : y + 2]
    neighbours = neighbours[~np.isnan(neighbours)]
    num_neighbours = np.sum(neighbours) - universe[x, y]
    # The rules of Life

    if cell == 1 and num_neighbours >= 4:
        return 0
    elif cell == 0 and num_neighbours == 0:
        return 1
    return cell
```

```python
def generation(universe):
    new_universe = np.copy(universe)
    # Apply the survival function to every cell in the universe
    for i in range(universe.shape[0]):
        for j in range(universe.shape[1]):
            new_universe[i, j] = survival(i, j, universe)
    return new_universe
```

```python
# Part 1
new_universe = np.copy(universe)

count = 0
new_count = 1
i = 0

while count != new_count:
    if i != 0:
        count = new_count
    new_universe = generation(new_universe)
    new_count = len(new_universe[new_universe == 1])
    i += 1
    logger.info(f'Gen {i}')

logger.info(f"Stablized after {i - 1} generation. Alived: {new_count}")
```

```python
def survival2(x, y, universe):
    cell = universe[x, y]
    if np.isnan(cell):  # Floor not change
        return cell
    occ = 0
    x_rest = x_end - x
    y_rest = y_end - y
    # Up
    for i in range(1, x):
        if universe[x-i, y] == 0:
            logger.debug(f"[UP] Found L at {x-i}, {y}")
            break
        if universe[x-i, y] == 1:
            logger.debug(f"[UP] Found # at {x-i}, {y}")
            occ += 1
            break
    # Down
    for i in range(1, x_rest):
        if universe[x+i, y] == 0:
            logger.debug(f"[DOWN] Found L at {x+i}, {y}")
            break
        if universe[x+i, y] == 1:
            logger.debug(f"[DOWN] Found # at {x+i}, {y}")
            occ += 1
            break
    # Left
    for i in range(1, y):
        if universe[x, y-i] == 0:
            logger.debug(f"[LEFT] Found L at {x}, {y-i}")
            break
        if universe[x, y-i] == 1:
            logger.debug(f"[LEFT]Found # at {x}, {y-i}")
            occ += 1
            break
    # Right
    for i in range(1, y_rest):
        if universe[x, y+i] == 0:
            logger.debug(f"[RIGHT]Found L at {x}, {y+i}")
            break
        if universe[x, y+i] == 1:
            logger.debug(f"[RIGHT] Found # at {x}, {y+i}")
            occ += 1
            break
    # Up left
    for i in range(1, min(x, y)):
        if universe[x - i, y - i] == 0:
            logger.debug(f"[UP LEFT] Found L at {x-i}, {y-i}")
            break
        if universe[x - i, y - i] == 1:
            logger.debug(f"[UP LEFT] Found # at {x-i}, {y-i}")
            occ += 1
            break
    # Up right
    for i in range(1, min(x, y_rest)):
        if universe[x - i, y + i] == 0:
            logger.debug(f"[UP RIGHT] Found L at {x-i}, {y+i}")
            break
        if universe[x - i, y + i] == 1:
            logger.debug(f"[UP RIGHT] Found # at {x-i}, {y+i}")
            occ += 1
            break
    # Down left
    for i in range(1, min(x_rest, y)):
        if universe[x + i, y - i] == 0:
            logger.debug(f"[DOWN LEFT] Found L at {x+i}, {y-i}")
            break
        if universe[x + i, y - i] == 1:
            logger.debug(f"[DOWN LEFT] Found # at {x+i}, {y-i}")
            occ += 1
            break
    # Down right
    for i in range(1, min(x_rest, y_rest)):
        if universe[x + i, y + i] == 0:
            logger.debug(f"[DOWN RIGHT] Found L at {x+i}, {y+i}")
            break
        if universe[x + i, y + i] == 1:
            logger.debug(f"[DOWN RIGHT] Found # at {x+i}, {y+i}")
            occ += 1
            break
    logger.debug(f'Before: {cell}')
    if cell == 1 and occ >= 5:
        cell = 0
    elif cell == 0 and occ == 0:
        cell = 1
    logger.debug(f'After: {cell}')
    return cell
```

```python
def generation2(universe):
    new_universe = np.copy(universe)
    # Apply the survival function to every cell in the universe
    for i in range(universe.shape[0]):
        for j in range(universe.shape[1]):
            new_universe[i, j] = survival2(i, j, universe)
    return new_universe
```

```python
# Part 2
new_universe = np.copy(universe)

count = 0
new_count = 1
i = 0

while count != new_count:
    logger.debug(new_universe)
    if i != 0:
        count = new_count
    new_universe = generation2(new_universe)
    new_count = len(new_universe[new_universe == 1])
    i += 1
    logger.info(f'Gen {i}')

logger.info(f"Stablized after {i - 1} generation. Alived: {new_count}")
```
